## Load Data

```python
sample_submission = pd.read_csv("data/sample_submission.csv")
sample_submission
```

## Creating Random Submission 👀
In below cell, we are going to create a random labels to submit our predictions.

```python
predictions = []


for n in range(0, sample_submission.shape[0]):

  moves = []

  no_of_moves = random.randint(5, 10)

  for i in range(no_of_moves):

      # Crating random location of piece
      previous_location = random.choice(string.ascii_lowercase[:8])+random.choice(string.digits[1:9])
      new_location = random.choice(string.ascii_lowercase[:8])+random.choice(string.digits[1:9])

      prediction = ''.join(previous_location+new_location)

      moves.append(prediction)

  predictions.append(" ".join(moves))
```

```python
sample_submission['label'] = predictions
sample_submission
```

## Save the prediction to csv

```python
sample_submission.to_csv("submission.csv", index=False)
```
